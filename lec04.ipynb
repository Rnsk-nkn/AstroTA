```python
import matplotlib . pyplot as plt
import math
ld = open ("dataLec03/hipparcos.dat")
lines = ld.readlines ()
ld.close ()
x = []
y = []
for line in lines :
    vapp = float ( line [129:136])
    bv = float ( line [152:158])
    d=1000/float(line[43:50])
    v=vapp+5-5*math.log10(d)
    if bv != 0.:
        x.append (bv)
        y.append (v)

ld2 = open ("dataLec04/praesepe.dat")
lines2 = ld2.readlines ()
ld2.close ()
xs = []
ys = []
for line in lines2 :
    vapp = line [12:19]
    bv = line [20:27]
    if vapp != "       " and bv != "       ":
        xs.append (float(bv))
        ys.append (float(vapp))
plt. xlim ( -0.5 ,2.5)
plt. ylim (16 , -2)
plt.xlabel('mB-mV')
plt.ylabel('mV, MV')
plt.scatter(x,y,s=1,label="hipparcos")
plt.scatter(xs,ys,s=1,label="praesepe")
plt.legend(bbox_to_anchor=(0.99,0.99),loc='upper right',fontsize=12)
plt.show()
```
