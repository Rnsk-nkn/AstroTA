```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# シミュレーションパラメータ
num_points = 100  # 点の初期数
num_iterations = 100  # シミュレーションの反復回数

# データの初期化
points = np.random.rand(num_points, 2)  # ランダムな位置に点を配置

# プロットの初期化
fig, ax = plt.subplots()
scatter = ax.scatter(points[:, 0], points[:, 1], color='blue', alpha=0.6)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Simulation Iteration: 0')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# フレーム更新の関数
def update(frame):
    # 点の増殖と移動
    num_new_points = np.random.randint(1, 4)  # 新たに生成する点の数
    new_points = np.random.rand(num_new_points, 2)  # ランダムな位置に新たな点を生成
    op=points
    points = np.concatenate([op, new_points])  # 新たな点を既存の点と結合
    points += np.random.randn(num_points + num_new_points, 2) * 0.02  # 点をランダムに移動

    # プロットの更新
    scatter.set_offsets(points)
    ax.set_title('Simulation Iteration: {}'.format(frame + 1))

# アニメーションの作成
animation = FuncAnimation(fig, update, frames=num_iterations, interval=200, repeat=False)

# アニメーションの保存
animation.save('simulation_animation.gif', writer='pillow')

plt.show()
```
