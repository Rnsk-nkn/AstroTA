```python
import matplotlib . pyplot as plt
import math
g = 10
dt = 0.01
for D in [15,30,45,60]:
    t=0
    x = 0
    y = 0
    v0 = 20
    X=[]
    Y=[]
    print(D)
    vx =  v0*math.cos(math.radians(D))
    vy = v0*math.sin(math.radians(D))
    tend = (2*vy)/g
    while t < tend: 
        X.append(vx*t)
        Y.append(vy*t-(1/2)*g*t**2)
        #x=vx*t
        #y=vy*t-(1/2)*g*t**2
        #print(x,y)
        t += dt
    plt. xlim ( 0 ,45)
    plt. ylim (0 , 20)
    plt. plot (X,Y)
plt. show ()
```
