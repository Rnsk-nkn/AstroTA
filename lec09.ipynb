```python
import numpy as np#new
import matplotlib.pyplot as plt#new
import random
import time
#tstart=time.time()
class star ():
  def __init__ ( self , name , posx , posy ):
    self.name = name
    self.posx = posx
    self.posy = posy
    self.nnbs = None #new最近傍

random.seed (1)
nstar = 100
stars = []
for i in range ( nstar ):
  xtemp = random . random ()
  ytemp =random . random ()
  stars.append ( star (i , xtemp , ytemp ))

#星描画
x = []
y = []
for i in range ( nstar ):
  xtemp = stars [ i ]. posx
  ytemp = stars [ i ]. posy
  x . append ( xtemp )
  y . append ( ytemp )
plt . scatter (x , y )
#最近傍の恒星の探査

tstart=time.time()
for i in range ( nstar ):
  r2min =1e+10
  for j in range ( nstar ):
    if stars [ j ]. name == stars [ i ]. name :
      continue #for文最初に戻る
    r2 = ((stars [ i ]. posx-stars [ j ]. posx)**2+(stars [ i ]. posy-stars [ j ]. posy)**2)#rootは省略**0.5 # stars [ i ] と stars [ j ] の 距 離 の ２ 乗 を 計 算
    if r2 < r2min :
      r2min = r2 # 最 小 距 離 の 更 新
      stars [ i ]. nnbs = stars [ j ]

tend=time.time()
#最近傍の描写
for i in range ( nstar ):
  xi = stars [ i ]. posx # stars [ i ] の x 座 標
  yi = stars [ i ]. posy # stars [ i ] の y 座 標
  dx = (stars [ i ].nnbs. posx-stars [ i ]. posx )# stars [ i ] の x 座 標 と そ の 最 近 傍 の 恒 星 の x 座 標 の 差 #new引き算
  dy = (stars [ i ].nnbs. posy-stars [ i ]. posy ) # stars [ i ] の y 座 標 と そ の 最 近 傍 の 恒 星 の y 座 標 の 差
  plt . arrow ( xi , yi , dx , dy , width =0.005 , length_includes_head = True )
plt.plot()
#tend=time.time()
print(nstar,tend-tstart)
```
